### 1. Investigating and Treating Null Values

```python
# Checking for NULL values in our dataframe

retail.isnull().sum()
```

The missing values in the CustomerID column refer to customers that do not log into an account associated with the company before placing an order. Hence, we can flag these guest logins with zero in our column.

```python
# Replacing NULL values in CustomerID column with 0, referring to Guest Login

retail.CustomerID = retail.CustomerID.fillna(0)
```

```python
# Investigating NULL values for Description column

retail[retail.Description.isnull()]
```

We can observe that the UnitPrice is zero for all rows, so we can drop them from our data frame as they will not affect our analysis.

```python
# Dropping rows where Description is NULL

print(retail[retail.Description.isnull()]['UnitPrice'].unique())

retail = retail[~retail.Description.isnull()]
```



---


### 2. Evaluating Data Quality

```python
# Converting columns to their relevant data types

retail = retail.astype({'InvoiceNo': str,	'StockCode' : str, 'Description' : str,	'Quantity' : int,	'CustomerID' : int, 'Country' : str})
```

```python
# Checking to see if we have any negative UnitPrice values in our dataframe

retail[retail.UnitPrice < 0]
```

```python
# Checking corresponding rows for bad debts

retail[retail.UnitPrice == 11062.06]
```

```python
# Dropping these 3 rows as they do not hold relevant information for our analysis

retail = retail[(retail.UnitPrice != 11062.06) & (retail.UnitPrice != -11062.06)]
```

```python
# Checking where UnitPrice = 0

retail[retail.UnitPrice == 0]
```

```python
# Dropping rows where UnitPrice = 0

retail = retail[retail.UnitPrice > 0]
```

### Question 3: What is the proportion of returns and orders on a monthly basis?

We observe that December and January have the highest proportions of returns, about 12.9% and 17% respectively in comparison to the total orders. This proportion is significantly higher than the proportion of returns for the other months, where returns constitute of less than 5% of the total orders. This could be presumably because customers are returning holiday gifts that they purchased as a result of over-purchasing. Other reasons for this higher volume of returns could be faulty products, missing shipments or incomplete orders. We should also note that our data has 9 extra days in the month of December which could be another factor affecting these results.

```python
# Creating aggregate sales and return data by month

monthly_orders = retail[retail.IsReturn == 0].groupby('Month')['Quantity'].agg('sum').reset_index()

monthly_returns = retail[retail.IsReturn == 1].groupby('Month')['Quantity'].agg('sum').reset_index()

# Converting Return quantities to positive for plotting

monthly_returns.Quantity = abs(monthly_returns.Quantity)

# Plotting sales and returns by month

fig, ax = plt.subplots(1, 1, figsize = (7, 7))

plt.style.use('seaborn-v0_8')
plt.grid(False)
ax.bar('Month', 'Quantity', data = monthly_orders, color = 'b')
ax.bar('Month', 'Quantity', data = monthly_returns, color = 'r')
plt.legend(['Orders', 'Returns'], loc = [0.35, 0.8])
ax.set_xlabel("Month", fontsize = 12)
ax.set_ylabel("Quantity", fontsize = 12)
ax.yaxis.set_major_formatter('{x:,.0f}')
ax.set_title("Orders vs. Returns by months", fontsize = 16)

plt.show()
```

```python
# Calculating the proportion of returns for each month

returns_perc= pd.DataFrame(round(((monthly_returns.Quantity/(monthly_orders.Quantity + monthly_returns.Quantity))*100), 2))

returns_perc=returns_perc.set_index(np.arange(1,13)).reset_index()

returns_perc=returns_perc.rename(columns={'Quantity':'Percentage of returns', 'index':'Month'})

returns_perc['Percentage of returns']=returns_perc['Percentage of returns'].astype(str) +'%'

returns_perc
```

### Deep dive into the trends for return orders

In the first visualisation, we observe that out of total returns, a high volume of orders are returned during December and January, which could be presumably due to the holiday season around this time. People are possibly returning extra gifts that they might have bought which did not come to use. Other reasons for this higher volume of returns could be faulty products, missing shipments or incomplete orders. We should also note that our data has 9 extra days in the month of December which could be another factor affecting these results.

For the second visualisation, we get to see that returns in January and December make up for almost two-thirds of the total return orders (about 65%) for the given timeframe in the dataset.

```python
# Filtering the dataframe based on return orders
returns = retail[retail.IsReturn==1]

# Converting negative quantities to absolute values for ease in visualisation
returns = abs(returns.groupby('Month')['Quantity'].sum()).reset_index()

# Creating a perc column to calculate percentage of returns each month as compared to the total
returns["Proportion"] = returns.Quantity/returns.Quantity.sum()*100
```

```python
# Plotting returns for each month

fig, ax = plt.subplots(1, 1, figsize = (8, 5))

ax.bar('Month', 'Quantity', data=returns)
ax.set_xlabel("Month", fontsize = 12)
ax.set_ylabel("Quantity", fontsize = 12)
plt.grid(False)
ax.yaxis.set_major_formatter('{x:,.0f}')
plt.title("Proportion of Returns each month", fontsize = 17)
for ind, r in returns.iterrows():
  plt.text(r['Month'], r['Quantity'], f'{r["Proportion"]:.1f}%', ha='center', va='bottom')
plt.show()
```

```python
# Plotting a pie chart to show distribution of returns in January and December as compared to other months

jan = returns.Quantity[0]
dec = returns.Quantity[11]
others = returns.Quantity.sum() - jan - dec
months = pd.Series([jan, dec, others])

plt.pie(months, labels=['January', 'December', 'Other Months'], autopct='%1.1f%%', colors=['grey', 'pink', 'orange'])
plt.show()
```

### Question 4: Which products are most frequently returned?


From the chart we can observe that some of the most frequently returned products include products like 'REGENCY CAKESTAND 3 TIER', 'JAM MAKING SET WITH JARS' and 'SET OF 3 CAKE TINS PANTRY DESIGN'.

From Question 5, it can also be observed that out of the top 10 most returned products, 5 of them are also part of the most purchased products. However, the ratio of returns to total orders for these products is low (less than 10%) and therefore is not a cause for concern for the company.




```python
# Creating a dataframe of rows that have IsReturn is equal to 1, then grouping them by description to find the most returned products
returned_products = retail[retail["IsReturn"] == 1]
return_counts = returned_products.groupby('Description')['IsReturn'].count().reset_index()

# Arranging them in descending order
returned_products_sorted = return_counts.sort_values(by='IsReturn', ascending=False).head(14)

# After taking a look at the list, we had to remove the following 4 descriptions as they are not products but accounting entries
excluded_descriptions = ['Manual', 'POSTAGE', 'Discount', 'SAMPLES']
filtered_returned_products_sorted = returned_products_sorted[~returned_products_sorted['Description'].isin(excluded_descriptions)]
```

```python
# Plotting the graph

plt.figure(figsize=(8, 5))
sns.set(style="whitegrid")
sns.barplot(x='IsReturn', y='Description', data=filtered_returned_products_sorted, palette="Set3")

# Setting labels and title

plt.xlabel("Number of Returns", fontsize=12)
plt.ylabel("Product Description", fontsize=12)
plt.grid(False)
plt.title("Top 10 Returned Products", fontsize=16)
plt.show()
```

### Question 9: What are the possible reasons for sales of weeks 5 and 6 being abnormally low?

A possible reason for this dip in sales in these weeks could be the days from 24th December, 2010 till 3rd January, 2011 when the company was not operational, and thus, did not generate any sales. Another possible reason could be that it is just a case of bad or missing data due to data warehousing issues at the company. Both reasons are plausible, but we will assume the former for the purposes of our analysis.

```python
# Adding a date column to check if there are any missing dates in the dataset

missing_dates = retail.groupby('InvoiceDate')[['Quantity', 'Sales']].agg('sum').reset_index().sort_values(['Quantity', 'InvoiceDate'])
missing_dates['Date']= missing_dates.InvoiceDate.dt.date

# We observe a pattern of a missing day each week, and a set of missing days from 24th Dec, 2010 till 3rd Jan, 2011

missing_dates.groupby('Date')[['Quantity', 'Sales']].agg(sum).reset_index().head(25)
```

```python
# Adding a column for days to investigate the pattern of missing days in each week

missing_days = missing_dates
missing_days['Day']= missing_days.InvoiceDate.dt.strftime('%a')

# Aggregating quantity and sales on 'Day' to get sales on unique days of the week

missing_days.groupby('Day')[['Quantity', 'Sales']].agg(sum)
```

From the above table, we also observe that Saturdays are missing in the entire dataset for each week. It seems that the company does not operate or process transactions on Saturdays.

### Question 10: Is there a trend of increased buying among consumers during different times of day as well?

As observed in the visualisation below, it can be inferred that the time slot of 12 PM to 6 PM is the most popular time of the day for people to place their orders. This makes up for an average of about two-thirds of the entire transactions that take place on a daily basis.

It is also to be noted that no orders take place between 8 PM and 6 AM in our data. This is due to the fact that the company does not process any orders during this time frame. This also affects the share of Evening orders (2%) as it only captures 2 hours of orders in comparison to 6 hours each for Morning and Afternoon.


```python
# Creating time of day bins for our data

time_bins = [0, 6, 12, 18, 24]

time_labels = ['Night', 'Morning', 'Afternoon', 'Evening']

retail['Time of Day'] = pd.cut(retail.InvoiceDate.dt.hour, bins = time_bins, labels = time_labels, right = False)
```

```python
# Checking performace of the company by time of day

time_of_day = retail.groupby('Time of Day').agg({'Sales' : 'sum', 'Quantity' : 'sum', 'InvoiceNo' : lambda x: len(x.unique()), 'IsReturn' : 'sum'}).reset_index()
time_of_day
```

```python
# Investigating why there is no data for night time. We found that no orders were processed between 8 PM and 6 AM

print(retail.InvoiceDate.dt.hour.min())
print(retail.InvoiceDate.dt.hour.max())
```

```python
# Plotting proportion of transactions by time of day

plt.style.use('fivethirtyeight')
plt.pie(time_of_day.InvoiceNo[1:], labels = time_of_day['Time of Day'][1:], autopct='%1.1f%%');
plt.title('Percentage of Transactions by Time of Day', fontsize = 14)
plt.show()
```

### Question 13: Comparison of wholesale and individual retail sales each month

On a monthly basis, the wholesale sales are about one-fourth of individual sales. This shows that the target consumers of the company include people who buy these gifts in reasonable amounts rather than in bulk. We can also observe that bulk sales increase during the months of November and December, reaching almost £400,000 during December. A possible reason could be that middlemen buy these products in bulk to resell them at a profit during the holiday season. Alternatively, it could be orders by other companies for their employees or clients.


```python
# Segregating the dataset into individual orders and wholesale orders depending on the quantity that was ordered

wholesale = retail[(retail.IsReturn == 0) & (retail.Quantity >= 100)]
individual_orders =  retail[(retail.IsReturn == 0) & (retail.Quantity < 100)]
```

```python
# Plotting a bar chart to visualise sales through wholesale and individual orders throughout the year

w = wholesale.groupby('Month')['Sales'].sum().reset_index()
i = individual_orders.groupby('Month')['Sales'].sum().reset_index()
n= len(w.Sales)

blue_bar= list(w.Sales)
orange_bar= list(i.Sales)

ind= np.arange(n)

fig, ax = plt.subplots(1, 1, figsize = (10, 6))

ax.bar(x = ind, height = blue_bar, width = 0.3, label='Wholesale', color='orange')
ax.bar(x = ind+ 0.3, height = orange_bar, width = 0.3, label='Individual', color='green')
ax.set_xticks(ticks = ind+(0.3/2), labels = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
ax.set_title('Monthly sales through wholesale and individual orders', fontsize = 16)
ax.yaxis.set_major_formatter('£{x:,.0f}')
ax.set_ylabel('Sales', fontsize = 14)
ax.set_xlabel('Month', fontsize = 14, labelpad = 10)
ax.legend(labels=['Wholesale', 'Individual'], loc='upper center')
plt.grid(False)

plt.show()
```

### Question 14: Who are the top customers in terms of revenue and number of transactions?

Our 10 most loyal customers account for about 5.6% of the total transactions and ~14% of the total sales.

As seen in the charts below, out of the 4372 customers that the company serves throughout the year, customers with IDs 14646 and 18102 account for over £500,000 in sales, which makes up for about 5% of the total revenue.

Further, it can also be observed that customers with IDs 14911 and 12748 make up for about 2% of the total number of transactions taking place in our dataset.

Some of the customers that contribute significantly to both sales and transactions are customers with IDs 14911, 15311, and 14646.

```python
# Creating a data frame to find the most loyal customers

top_customers_transactions = retail.groupby('CustomerID').agg({'Sales' : 'sum', 'Quantity' : 'sum', 'InvoiceNo' : lambda x: len(x.unique()), 'IsReturn' : 'sum'}).reset_index().sort_values('InvoiceNo', ascending = False)
top_customers_sales = retail.groupby('CustomerID').agg({'Sales' : 'sum', 'Quantity' : 'sum', 'InvoiceNo' : lambda x: len(x.unique()), 'IsReturn' : 'sum'}).reset_index().sort_values('Sales', ascending = False)
```

```python
# Excluding guest logins from our analysis

top_customers_transactions = top_customers_transactions[top_customers_transactions.CustomerID > 0]
top_customers_sales = top_customers_sales[top_customers_sales.CustomerID > 0]
```

```python
# Plotting the top 10 most loyal customers by transactions and sales

fig, axs = plt.subplots(1, 2, figsize = (12, 5))
plt.style.use('fivethirtyeight')

# By transactions

sns.barplot(x = top_customers_transactions['InvoiceNo'][0:10], y = top_customers_transactions['CustomerID'][0:10], order = top_customers_transactions['CustomerID'][0:10], ax = axs[0], orient = 'h');
axs[0].set_xlabel('Transactions', fontsize = 12, labelpad = 10)
axs[0].set_ylabel('CustomerID', fontsize = 12, labelpad = 10)
axs[0].set_title('Most loyal customers by Transactions', fontsize = 16)

# By sales

sns.barplot(x = top_customers_sales['Sales'][0:10], y = top_customers_sales['CustomerID'][0:10], order = top_customers_sales['CustomerID'][0:10], ax = axs[1], orient = 'h');
axs[1].set_xlabel('Sales', fontsize = 12, labelpad = 10)
axs[1].set_ylabel('CustomerID', fontsize = 12, labelpad = 10)
axs[1].set_title('Most loyal customers by Sales', fontsize = 16)
axs[1].xaxis.set_major_formatter('£{x:,.0f}')
axs[1].tick_params(axis = 'y', top = False)

plt.show()
```

## **Data Challenges**

The dataset commences on December 1, 2010, and concludes on December 9, 2011. This has an impact on the analysis of monthly data, as it extends the duration of December by 9 additional days. In addition, our dataset contains entries for countries like 'Channel Islands', 'European Community' and 'Unspecified' for which ISO country codes do not exist. Consequently, we found it necessary to exclude these countries from our code when creating interactive map plot. Notably, this exclusion did not significantly affect our overall results, as sales from these particular countries were relatively low.

Furthermore, it's worth noting that the dataset exhibits a noticeable absence of entries from December 24, 2010 to January 3, 2011. This data gap could be attributed to various factors, such as the company's potential non-operation during the holiday season, bad data issues, or missing information. Lastly, it's important to acknowledge that the dataset includes a single Customer ID for all guest logins, a detail that we took into account throughout our analysis.
